```python
# Ellipsoid Function
def ellipsoid(x, alpha=1000):
    d = len(x)
    alpha_vec = np.fromfunction(lambda i: alpha**((i/(d-1))), (d,), dtype=int)
    return alpha_vec @ x**2

def ellipsoid_2d(x, y, alpha=1000):
  return x**2 + alpha*y**2
  
def gradient_ellipsoid(x,alpha=1000):
    d = len(x)
    weight_vec=np.zeros(d)
    for i in range(d):
      if d == 1:
        weight_vec[i]=np.inf
        continue
      weight_vec[i]=alpha**(i/(d-1))
    return 2*weight_vec*x

def hessian_ellipsoid(x,alpha=1000):
    d=len(x)
    weight_vec=np.zeros(d)
    for i in range(d):
        weight_vec[i]=alpha**(i/(d-1))
    return np.diag(2*weight_vec)

def banana(x):
    return (1 - x[0])**2 + (100 * (x[1] - (x[0]**2))**2)

# Log-Ellipsoid
def logellipsoid(x, epsilon=10**-16):
    return np.log1p(ellipsoid(x) / epsilon)

def logell_2d(x, y, epsilon=10**-16):
  return np.log(epsilon + ellipsoid_2d(x,y))

def logellipsoid_gradient(x, epsilon=1**-16):
    return gradient_ellipsoid(x)*(2/(epsilon+ellipsoid(x)))

def hessian_logellipsoid(x,epsilon=1**-16):
    matrix1=-(epsilon+ellipsoid(x))**(-2)*gradient_ellipsoid(x).reshape(-1,1)@gradient_ellipsoid(x).reshape(1,-1)
    matrix2=hessian_ellipsoid(x)/(epsilon+ellipsoid(x))
    return matrix1+matrix2

# h function with derivatives 
def h(x, q = 10**8):
    return (np.log(1 + np.exp(-np.fabs(q*x))) + np.maximum(q*x, np.zeros_like(x))) / q

def dh_dx(x,q=1e8):
    if x >= 0:
        return 1 / (1 + np.exp(-q * x))
    else:
        return np.exp(q*x) / (1 + np.exp(q * x))

def grad_h(x, q = 10**8):
    return np.vectorize(dh_dx)(x)

# f4 attr. sector function
def f4(x):
    values = np.array([h(k)+100*h(-k) for k in x])
    return np.sum(values)

def f4_2D(x,y):
  return h(x)+100*h(-x)+h(y)+100*h(-y)

def gradient_f4(x,q=1e8):
    return grad_h(x) - (100 * grad_h(-x))

# f5 attr. sector function 2
def f5(x):
  values = np.array([h(k)**2 + 100 * h(-k)**2 for k in x])
  return np.sum(values)

def f5_2D(x,y):
  return h(x)**2+100*h(-x)**2 + h(y)**2+100*h(-y)

def gradient_f5(x,q=1e8):
  return 2 * h(x) * grad_h(x) - (200 * h(-x) * grad_h(-x))

def dh_dx(x,q=1e8):
    if x<0:
        return np.exp(q*x)/(1+np.exp(q*x))
    return 1+dhminus_dx(x)

def dhminus_dx(x,q=1e8):
    if x>=0:
        return -np.exp(-q*x)/(1+np.exp(-q*x))
    return dh_dx(x)-1

# second derivatives (Hessian) for f4 and f5 
def d2h_dx2(x,q=1e8):
    if x<0:
        return q*np.exp(q*x)*(1+np.exp(q*x))**(-2)
    return q*np.exp(-q*x)*(1+np.exp(-q*x))**(-2)

def d2hminus_dx2(x,q=1e8):
    return d2h_dx2(x)

def hessian_f4(x,q=1e8):
    part1=np.array(list(map(d2h_dx2,x)))
    part2=np.array(list(map(d2hminus_dx2,x)))
    return np.diag(part1+100*part2)

def hessian_f5(x,q=1e8):
    d2h_dx2_values=np.array(list(map(d2h_dx2,x)))
    d2hminus_dx2_values=np.array(list(map(d2hminus_dx2,x)))
    hx_values=np.array(list(map(h,x)))
    hminusx_values=np.array(list(map(h,-x)))
    dh_dx_values=np.array([dh_dx(x_i) if x_i<0 else 1+dhminus_dx(x_i) for x_i in x])
    dhminus_dx_values=np.array([dhminus_dx(x_i) if x_i>=0 else dh_dx(x_i)-1 for x_i in x])
    part1=2*np.square(dh_dx_values)+2*hx_values*d2h_dx2_values
    part2=-200*np.square(dhminus_dx_values)-200*hminusx_values*d2hminus_dx2_values
    return np.diag(part1-part2)

def gradient_banana(x):
    item1=400*x[0]**3-400*x[0]*x[1]+2*x[0]-2
    item2=200*x[1]-200*x[0]**2
    return np.array([item1,item2])
def hessian_banana(x):
    dx1dx1=1200*x[0]**2-400*x[1]+2
    dx1dx2=-400*x[0]
    dx2dx1=-400*x[0]
    dx2dx2=200
    return np.array([[dx1dx1,dx1dx2],
                    [dx2dx1,dx2dx2]])
```

```python
def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)
```

```python
def m(f, B, g, x, p):
    return f(x) + np.dot(g(x).transpose(), p) + 0.5 * (np.dot(np.dot(p.transpose(), B), p))
```

```python
def p(B, lam, g, delta):
    if lam != 0:
        #assert is_pos_def(B + lam*np.identity(B.shape[0]))
        p = np.dot(-np.linalg.inv(B + lam*np.identity(B.shape[0])), g)
    else:
        p = -np.dot(np.linalg.inv(B), g)
    #assert np.linalg.norm(p) <= delta + 1e-10
    return p

def solve_subproblem(g, B, delta, max_iter = 100000, lam_0=0, lam_1=100):
    # Use bijection algorithm to find pk that solves subproblem 4.7
    iter_count = 0
    assert delta > 0
    random_number = random.uniform(-100, 100)
    if np.linalg.norm(p(B, random_number, g, delta)) > delta:
        iter_count = 0
        lambda_0 = random_number
        lam_0 = lambda_0
        tmp = random.uniform(lambda_0, lambda_0 + 100)
        while 'lambda_1' not in locals() and iter_count <= max_iter:
            tmp += (1**iter_count)
            iter_count += 1
            if np.linalg.norm(p(B, tmp, g, delta)) < delta:
                lambda_1 = tmp
                lam_1 = lambda_1
    else:
        iter_count = 0
        lambda_1 = random_number
        lam_1 = lambda_1
        tmp = random.uniform(lambda_1 - 100, lambda_1)
        while 'lambda_0' not in locals() and iter_count <= max_iter:
            tmp += (1**iter_count)
            iter_count += 1
            if np.linalg.norm(p(B, tmp, g, delta)) > delta:
                lambda_0 = tmp  
                lam_0 = lambda_0
    if not ('lambda_0' in locals() and 'lambda_1' in locals()):
        lambda_0, lambda_1 = lam_0, lam_1
    lambda_prime = (lambda_0 + lambda_1) * 0.5
    while (np.linalg.norm(p(B, lambda_0, g, delta)) <= delta or np.linalg.norm(p(B, lambda_1, g, delta)) >= delta) and iter_count <= max_iter:
        lambda_prime = (lambda_0 + lambda_1) * 0.5
        if np.linalg.norm(p(B, lambda_prime, g, delta)) > delta:
            lambda_0 = lambda_prime
        else:
            lambda_1 = lambda_prime
        iter_count += 1
    print('lambda_prime: ', lambda_prime)
    return p(B, lambda_prime, g, delta), lam_0, lam_1
            
    

def trust_region(f, g, hess, x, delta_hat = 10000, eta = 0.125, delta=None, tmp=1e12, tol=1e-5, max_iter = 100):
    iter_count = 0
    gradient_norm_lst = []
    gradient_norm_lst.append(np.log(np.linalg.norm(g(x))))
    delta = 1
    delta_lst = [delta]
    print('delta: ', delta)
    B = hess(x)
    while not is_pos_def(B):
        B = B + 1 * np.identity(B.shape[0])
    while iter_count <= max_iter and np.linalg.norm(g(x)) >= tol:
        try:
            p, lam_0, lam_1 = solve_subproblem(g=g(x), B=B, delta=delta, lam_0=lam_0, lam_1=lam_1)
        except:
            p, lam_0, lam_1 = solve_subproblem(g=g(x), B=B, delta=delta)
        print('p: ', p)
        rho = abs(f(x) - f(x+p)) / abs(f(x) - m(f=f, B=B, g=g, x=x, p=p))
        if rho < 0.25:
            delta *= 0.25
            print('delta: ', delta)
        elif rho > 0.75 and np.linalg.norm(p) > delta:
            delta = min(2*delta, delta_hat)
            print('delta: ', delta)
        else:
            print('delta: ', delta)
        delta_lst.append(delta)
        print('rho: ', rho)
        if rho > eta:
            x = x + p
            print('x: ', x)
            tmp = np.linalg.norm(g(x))
        else:
            print('x: ', x)
        gradient_norm_lst.append(np.log(np.linalg.norm(g(x))))
        iter_count += 1
    print('final x: ', x)
    print('gradient norm: ', np.linalg.norm(x))
    plt.plot(gradient_norm_lst)
    plt.title('First Attractive-sector')
    plt.xlabel('Iteration')
    plt.ylabel('Log of Gradient Norm')
    plt.show()
    plt.plot(delta_lst)
    plt.title('F5 Trust Region')
    plt.xlabel('Iteration')
    plt.ylabel('Delta')
    plt.show()
    return 
    
```
